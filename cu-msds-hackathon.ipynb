```python
## Are all entries of the same kingdom sampled from the same DNAtype?
## Answer: No, not all entries from the same kingdom use the same DNAtype as sample.
pd.pivot_table(df_train, values="id", index="DNAtype", columns="Kingdom", aggfunc="count")
```

```python
## Is the number of codons the same for each species when sampled from the same DNAtype?
mask = df_train.DNAtype == 0.0
display(df_train[mask].head())

## This shows that there are enough species that we don't have the same species under the same DNAtype
(df_train[mask].SpeciesID.value_counts() > 1).sum()
```

```python
## Of the same species (identified as having the same SpeciesID), are there multiple DNAtypes?
## Answer: Probably not because by just looking at the counts of different species, most of the species only have one instance in the training set.
df_train.SpeciesID.value_counts()
```

```python
## Do entries from the same Kingdom and same DNAtype have similar number of codons?
## Answer: No. The variance for each DNAtype and Kingdom is pretty high thus my hypothesis is that the Ncodons is heavily determined by the species
pd.pivot_table(df_train, values="Ncodons", index="DNAtype", columns="Kingdom", aggfunc=np.var)
```

```python
## Reduce the dimensions of the data by mapping codons to amino acids

## Create a codon to amino acid map from dataset
codon_amino_table = pd.read_csv("/kaggle/input/codons/Codons - Sheet1 (1).csv")
codon_amino_table = codon_amino_table.iloc[:, [0, 3]].set_index("codon").to_dict()
codon_amino_table = codon_amino_table["fullname"]

print(f"Amino Acid Table Length: {len(set(codon_amino_table.values()))}")

## Map and concat to create dimension-reduced training dataset
transposed_codons = df_train.iloc[:, 6:].T.reset_index().rename(columns={"index": "codon"})
amino_abbrev = transposed_codons["codon"].map(codon_amino_table).rename(index="AminoAcidAbbrev")
display(amino_abbrev)

transposed_codons = pd.concat([amino_abbrev, transposed_codons], axis=1)
transposed_codons = transposed_codons.drop("codon", axis=1)

codons = transposed_codons.groupby(by="AminoAcidAbbrev", axis=0).sum().T
codons = pd.concat([df_train.iloc[:, :6], codons], axis = 1)
display(codons)

## Save the dimension-reduced dataset to CSV
codons.to_csv("training_data_aminoAcid.csv", index=False)
```
